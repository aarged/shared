## Q1 - PLOTTING PANDA DATA FRAMES, MANIPULATING DATA AND PLOTTING
***


```python
import numpy as np # import libraries
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
df = [] # declare dataframe as an empty list
dr = pd.date_range('2016-01-01', 'today') # define date range
col = ('Column_1','Column_2','Column_3','Column_4','Column_5','Column_6') # establish number of columns
rows = len(dr) # determine length of date range for iterative loop
i = 0

while i < rows:
    d = np.random.uniform(low = 0, high = 1, size = 6) # create a row of random values between 0 and 1
    df.append(d) # append row to list
    i = i + 1


df = pd.DataFrame(df, columns = col, index = dr) # combine all components to complete the dataframe
df
```

```python
print("There are", rows, "rows in the dataframe", "\n")
df.head(30) # first 30 rows of data - output looks nicer without using print() command
```

```python
df_accumulate = df.cumsum() # calculate cumulative sum of each column
print(df.head(5)) 
print()
print(df_accumulate.head(5)) # compare results & verify values
```

```python
df_accumulate.plot(subplots=False, title="Cumulative Totals of Random Data", figsize=(10,8)) # all six columns in one graph
```

```python
df_accumulate.plot(subplots=True, title="Cumulative Totals of Random Data (plotted separately)",
        layout=(2, 3), figsize=(10,8)) # separate plots
```

## Q4 - CLEANING DATA
***


```python
mhdf.Gender.unique() # inspect all values for Gender.
mhdf['Gender'] = mhdf['Gender'].replace(['male','M','m','Make','Man','Mail','Malr','msle','Mal','maile', 'Male ', 'Cis Male', 'Male (CIS)', 'cis male', 'Cis Man', 'Guy (-ish) ^_^'], 'Male')
mhdf['Gender'] = mhdf['Gender'].replace(['female', 'F', 'f', 'Female ', 'Femake', 'woman', 'femail', 'Woman', 'Cis Female', 'cis-female/femme', 'Female (cis)'], 'Female')
# erroneous instances not specified in the question have not been corrected
mhdf.Gender.unique() # check output to ensure changes have been implemented
```

```python
mhdf_male = mhdf.loc[(mhdf.Gender == 'Male')]
print(mhdf_male.treatment.value_counts()) # check for erroneous values, obtain count to validate wedge size & labels
labels = 'Did not seek treatment', 'Sought treatment'


mhdf_male.treatment.value_counts().plot.pie(explode=[0.05,0.05],
                                            autopct='%1.1f%%', 
                                            figsize=(10,8), 
                                            title = 'Male Respondents Seeking Treatment',
                                            startangle=90,
                                            labels=['Did not seek treatment', 'Sought treatment'],
                                           shadow=True,
                                           rotatelabels=True)
```

```python
mhdf_fem = mhdf.loc[(mhdf.Gender == 'Female')]
print(mhdf_fem.treatment.value_counts()) 

mhdf_fem.treatment.value_counts().plot.pie(explode=[0.05,0.05],
                                            autopct='%1.1f%%', 
                                            figsize=(10,8), 
                                            title = 'Female Respondents Seeking Treatment',
                                            startangle=90,
                                            labels=['Sought treatment', 'Did not seek treatment'], # reversed order after validation
                                           shadow=True,
                                           rotatelabels=True)
```
